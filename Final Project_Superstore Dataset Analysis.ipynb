```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

# Data Loading And Initial Overview

```python
df = pd.read_csv(r"C:\Users\adhar\Downloads\Sample - Superstore.csv", encoding="latin1")
df.head()
df
```

```python
# Basic overview
print("Shape (rows, cols):", df.shape)

print("\nColumn data types:")
print(df.dtypes)

print("\nSummary statistics (numeric + categorical):")
print(df.describe(include='all').T)

print("\nMissing values per column:")
print(df.isna().sum())

print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Make column names neat
df.columns = df.columns.str.strip()

# Trim spaces inside text columns 
str_cols = df.select_dtypes(include="object").columns
df[str_cols] = df[str_cols].apply(lambda s: s.str.strip())
```

# Removing Duplicates

```python
# Remove duplicate rows
before = len(df)
df = df.drop_duplicates()
print(f"Removed {before - len(df)} duplicate rows")
```

# Correcting Datatypes

```python
# Fix dates and numbers
# Dates
for c in ['Order Date', 'Ship Date']:
    if c in df.columns:
        df[c] = pd.to_datetime(df[c], errors='coerce')

# Numbers
for c in ['Sales','Quantity','Discount','Profit','Shipping Cost']:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```

```python
df
```

# Handling Missing Values

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Missing values (top 10 columns):")
print(df.isna().sum().sort_values(ascending=False).head(10))
df.info()
```

# Creating Derived Columns

```python
# Derived Column 1 - Profit Margin (Profit / Sales)
df['Profit_Margin'] = (df['Profit'] / df['Sales']).round(2)
```

```python
#  Derived Column 2 - Sales per Quantity (unit price estimate)
df['Sales_per_Quantity'] = (df['Sales'] / df['Quantity']).round(2)
```

```python
df.info()
```

# Filtering Data

```python
#  1: Filter high value orders (Sales > 1000)
high_value_orders = df[df['Sales'] > 1000]
```

```python
high_value_orders.head(10)
```

```python
# 2: Aggregate total sales by Region
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
```
